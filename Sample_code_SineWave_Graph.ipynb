```python
#Sine wave for FFT
samples=250
f=3
x=np.arange(samples)
y1=np.sin(2*np.pi*f * (x/samples))
plt.figure()
plt.stem(x, y1, 'r')
plt.plot(x,y1)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```
